1. What is the difference between AI, ML, DL, and Data Science? Provide a

- Difference between AI, ML, DL, and Data Science

     Data Science is about collecting data, cleaning it, analyzing it, and finding meaningful insights.
     Machine Learning is used as a tool inside Data Science.
     The main purpose is to help businesses make better decisions using data.

2. Explain overfitting and underfitting in ML. How can you detect and prevent them?


  - Overfitting and Underfitting in Machine Learning
  
  -  Overfitting

     Overfitting happens when a model learns too much from the training data even the noise and random patterns.
     Because of this, the model performs very well on training data but poorly on test or new data.

  - How to Detect Overfitting

    - Training accuracy is very high but test accuracy is low.
    - Model becomes too complex.

  - How to Prevent Overfitting

    - Use more training data.
    - Use Dropout (in deep learning).
    - Perform Cross-Validation.



 -  Underfitting

     Underfitting happens when the model is too simple and cannot learn the patterns in data.
     Because of this, both training and test performance are poor.

 - How to Detect Underfitting

   - Low training accuracy and low test accuracy.
   -  Model predictions are very basic or inaccurate.
   - Model cannot capture the relationship between input and output.

 - How to Prevent Underfitting

   - Use a more complex model.
   - Train for more epochs/iterations.


3. How would you handle missing values in a dataset? Explain at least three
methods with examples?

- Handling Missing Values in a Dataset

     Missing values occur when some data points are not recorded.
     If not handled properly, they can reduce model accuracy or give wrong results.
     Here are three commonly used methods with clear examples:

   -  Deletion Method (Removing Missing Data)
   -  Remove Rows

      If only a few rows have missing values, we can delete those rows.

     Example:
     Dataset:

     Name	  Age	   Salary
      A	     25	   30,000
      B	—	28,000
      C	30	35,000

     Row of B has missing Age → delete this row.

     When to use

     When missing values are very few

     And deleting them won’t affect the dataset

     2) Imputation Method (Filling Missing Values)
     a) Mean Imputation

     For numerical columns, replace missing values with the mean.

     Example:
     Ages: 20, 25, 30, —
     Mean = (20 + 25 + 30) / 3 = 25
     So missing value becomes 25.

     b) Median Imputation

     Useful when data has outliers.

     Example:
     Salaries: 20k, 21k, 22k, —, 5 lakh (outlier)
     Median = 22k → fill the missing value with 22k.

     c) Mode Imputation

     For categorical data.

     Example:
     City: Delhi, Delhi, Jaipur, —
     Mode = Delhi → replace missing value with “Delhi”.

     3) Using Advanced Models
     a) KNN Imputation

     K-Nearest Neighbors fills missing values based on similar rows.

     Example:
     If a student’s score is missing, KNN finds other students with similar age & study hours and uses their average to fill the value.

     b) Regression Imputation

     Use a regression model to predict the missing value.

     Example:
     If “Salary” is missing, we can predict it using factors like Age, Education, and Experience.


4. What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

- Imbalanced Dataset

     An imbalanced dataset is a dataset where the classes are not represented equally.
     One class has many more samples than the other.
     Example:

     - Spam Classification

       - 950 normal emails

       - 50 spam emails
- Techniques to Handle Imbalanced Datasets  
  
   - Oversampling
     Theory:

     Oversampling means increasing the number of samples in the minority class so that both classes become balanced.
     It prevents the model from ignoring the minority class.

     Practical Example:

     Minority class (spam): 50 emails

     Oversample to 300 emails (duplicate or generate similar samples)

     Now both classes are roughly equal → model learns better


- Undersampling   
   
   Theory:

     Undersampling reduces the number of samples in the majority class.
     This makes both classes balanced, but it also removes some data.

   Practical Example:

     Majority class (normal): 950 emails

     Undersample to 300 emails

     Now both classes have equal frequency.


5. Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

- Feature scaling is important because different features in a dataset may have different ranges. If one feature has values like 2–10 and another has 1000–5000, then the model will give more importance to the larger numbers, even if they are not more important.

    - comparison of min-max scalling
      - 1. Min-Max Scaling (Normalization)
       
       Converts all values into a 0–1 range.

       Good for image data or cases where boundaries are required.

       Downside: Outliers can spoil the entire scaling.

      - Standardization (Z-score Normalization)

      Converts data so that mean = 0 and standard deviation = 1.

      Works better when the data is close to a normal distribution.

      Not easily affected by outliers.

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?


- Label Encoding

     Label Encoding is a technique that converts each category in a feature into a unique numeric label (0, 1, 2…).
     It is mainly used when the categorical values have a natural order (ordinal data) or when using tree-based models.
     However, it may create a false sense of ranking between categories if the data is not ordered.

- One-Hot Encoding

     One-Hot Encoding is a technique that converts categorical values into multiple binary columns, where each column represents one category with values 0 or 1.
     It is used when the data has no natural order (nominal data) and we want to avoid giving any numeric priority to categories.
     The drawback is that it can increase the number of columns in the dataset.
      
     - When to Prefer Which?

        - Label Encoding:
          Prefer it when categories are ordered or when using tree-based algorithms like Decision Tree, Random Forest, or XGBoost.

        - One-Hot Encoding:
          Prefer it when categories are unordered and when using linear models like Logistic Regression, Linear Regression, SVM, or KNN.

```python
#7 Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# Display first few rows
print("Dataset loaded successfully!")
print(df.head(), "\n")

# Step 3: Data Cleaning
# Drop duplicate entries
df.drop_duplicates(inplace=True)

# Drop rows with missing Rating values
df = df.dropna(subset=['Rating'])

# Keep only valid ratings (between 1 and 5)
df = df[(df['Rating'] >= 1) & (df['Rating'] <= 5)]

# Step 4: Handle missing or invalid categories
df = df.dropna(subset=['Category'])

# Step 5: Calculate average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

print("Average Rating per Category:\n")
print(category_ratings)

# Step 6: Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=category_ratings.values, y=category_ratings.index, palette="coolwarm")
plt.title("Average App Rating by Category", fontsize=14, fontweight='bold')
plt.xlabel("Average Rating")
plt.ylabel("App Category")
plt.xlim(3.0, 5.0)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 7: Insights
top_category = category_ratings.idxmax()
lowest_category = category_ratings.idxmin()

print(f"\n✅ Highest Rated Category: {top_category} ({category_ratings.max():.2f})")
print(f"❌ Lowest Rated Category: {lowest_category} ({category_ratings.min():.2f})")

print("\nPossible Reasons:")
print(f"- {top_category} apps likely provide educational or creative value with fewer bugs and ads.")
print(f"- {lowest_category} apps may face user dissatisfaction due to poor performance, privacy concerns, or intrusive ads.")
```

```python
#8 a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened? b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival? Dataset: https://github.com/MasteriNeuron/datasets.git

# -------------------------------------------------------------
# Titanic Survival Analysis
# (a) Survival vs Passenger Class (Pclass)
# (b) Survival vs Age Group (Children vs Adults)
# Dataset: https://github.com/MasteriNeuron/datasets.git
# -------------------------------------------------------------

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
df = pd.read_csv(url)

print("✅ Dataset loaded successfully!\n")
print(df.head(), "\n")

# Step 3: Basic cleanup
# Drop irrelevant columns (if any) and missing Age values for analysis
df = df.dropna(subset=['Age', 'Pclass', 'Survived'])

# Step 4: (a) Survival rate by Passenger Class
pclass_survival = df.groupby('Pclass')['Survived'].mean().sort_index()

print("=== (a) Survival Rate by Passenger Class ===")
print(pclass_survival, "\n")

# Visualization - Survival Rate by Class
plt.figure(figsize=(8,5))
sns.barplot(x=pclass_survival.index, y=pclass_survival.values, palette="viridis")
plt.title("Survival Rate by Passenger Class", fontsize=14, fontweight='bold')
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 5: (b) Survival rate by Age Group
# Create new column for Age Group
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

agegroup_survival = df.groupby('AgeGroup')['Survived'].mean()

print("=== (b) Survival Rate by Age Group ===")
print(agegroup_survival, "\n")

# Visualization - Survival Rate by Age Group
plt.figure(figsize=(6,5))
sns.barplot(x=agegroup_survival.index, y=agegroup_survival.values, palette="coolwarm")
plt.title("Survival Rate: Children vs Adults", fontsize=14, fontweight='bold')
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Step 6: Insights
highest_class = pclass_survival.idxmax()
lowest_class = pclass_survival.idxmin()

print(f"🏆 Highest survival rate: Class {highest_class} ({pclass_survival.max():.2f})")
print(f"⚠ Lowest survival rate: Class {lowest_class} ({pclass_survival.min():.2f})")

if agegroup_survival['Child'] > agegroup_survival['Adult']:
    print(f"\n👶 Children had a higher survival rate ({agegroup_survival['Child']:.2f}) than adults ({agegroup_survival['Adult']:.2f}).")
else:
    print(f"\n🧑 Adults had a higher survival rate ({agegroup_survival['Adult']:.2f}) than children ({agegroup_survival['Child']:.2f}).")

print("\n=== Analysis Summary ===")
print(f"- First-class passengers had the highest survival rate because they were located closer to lifeboats and prioritized during evacuation.")
print(f"- Third-class passengers had the lowest survival rate likely due to limited access to lifeboats and lower deck locations.")
print(f"- Children generally had better survival chances, reflecting the 'women and children first' policy during evacuation.")
```

```python
#Question 9: Flight Price Prediction Dataset a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window. b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why? Dataset: https://github.com/MasteriNeuron/datasets.git
# -------------------------------------------------------------
# Flight Price Prediction Analysis
# (a) Price vs Days Left
# (b) Price Comparison across Airlines (e.g., Delhi–Mumbai)
# Dataset: https://github.com/MasteriNeuron/datasets.git
# -------------------------------------------------------------

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 2: Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

print("✅ Dataset loaded successfully!\n")
print(df.head(), "\n")

# Step 3: Basic cleanup
# Drop missing or invalid prices
df = df.dropna(subset=['price', 'days_left', 'airline', 'source_city', 'destination_city'])

# Convert Price to numeric if necessary
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])

# Step 4: (a) Analyze how flight prices vary with days left
plt.figure(figsize=(10,6))
sns.scatterplot(x='days_left', y='price', data=df, alpha=0.6, color='teal')
sns.lineplot(x='days_left', y='price', data=df.groupby('days_left')['price'].mean().reset_index(), color='red', label='Average Price')
plt.title("Flight Price vs Days Left to Departure", fontsize=14, fontweight='bold')
plt.xlabel("Days Left Until Departure")
plt.ylabel("Flight Price (INR)")
plt.grid(alpha=0.4)
plt.legend()
plt.show()

# Compute average price per day left
avg_price_by_days = df.groupby('days_left')['price'].mean().sort_index()

# Identify price surge patterns
price_diff = avg_price_by_days.diff()
surge_points = price_diff[price_diff > avg_price_by_days.std()].index.tolist()

print("=== (a) Price Variation with Days Left ===")
print(avg_price_by_days.head(10), "\n")
print(f"⚠ Detected potential exponential price surges close to these days before departure: {surge_points}")

# Best booking window (based on lowest average prices)
best_window = avg_price_by_days.idxmin()
print(f"💡 Recommended Booking Window: Around {best_window} days before departure for best prices.\n")

# Step 5: (b) Compare prices across airlines for the same route (Delhi → Mumbai)
route_df = df[(df['source_city'].str.lower() == 'delhi') & (df['destination_city'].str.lower() == 'mumbai')]

if not route_df.empty:
    airline_prices = route_df.groupby('airline')['price'].mean().sort_values()

    print("=== (b) Average Prices by Airline (Delhi → Mumbai) ===")
    print(airline_prices, "\n")

    # Visualization
    plt.figure(figsize=(10,5))
    sns.barplot(x=airline_prices.values, y=airline_prices.index, palette="coolwarm")
    plt.title("Average Flight Price by Airline (Delhi–Mumbai)", fontsize=14, fontweight='bold')
    plt.xlabel("Average Price (INR)")
    plt.ylabel("Airline")
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    cheapest_airline = airline_prices.idxmin()
    premium_airline = airline_prices.idxmax()

    print(f"🪙 Cheapest Airline (Delhi–Mumbai): {cheapest_airline} (Avg ₹{airline_prices.min():.0f})")
    print(f"💎 Premium Airline (Delhi–Mumbai): {premium_airline} (Avg ₹{airline_prices.max():.0f})")

    print("\nPossible Reasons:")
    print(f"- {cheapest_airline} likely offers competitive or budget fares, minimal amenities.")
    print(f"- {premium_airline} offers full-service experience, better timing, or brand premium.")
else:
    print("⚠ No Delhi–Mumbai route found in the dataset. Try another route.")
```

```python
#Question 10: HR Analytics Dataset a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary). b). Are employees with more projects more likely to leave? Dataset: hr_analytics
# -------------------------------------------------------------
# HR Analytics - Employee Attrition Analysis
# (a) Correlation and Key Drivers of Attrition
# (b) Relationship between Number of Projects and Attrition
# -------------------------------------------------------------

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

print("✅ Dataset loaded successfully!\n")
print(df.head(), "\n")

# Step 3: Basic cleaning
# Drop duplicates and handle missing values if any
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Step 4: Convert categorical columns where needed
# Convert 'left' to binary (1 for left, 0 for stayed)
df['left'] = df['left'].apply(lambda x: 1 if x == 1 else 0) # Assuming 'left' is already 0 or 1, but ensuring type

# Step 5: (a) Find key factors correlated with attrition
# Select numeric columns for correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()['left'].sort_values(ascending=False)

print("=== (a) Correlation of Variables with Employee Attrition ===")
print(corr, "\n")

# Step 6: Visualization of key drivers
top_factors = corr[1:6]  # Top 5 excluding 'left' itself

plt.figure(figsize=(8,5))
sns.barplot(x=top_factors.values, y=top_factors.index, palette="coolwarm")
plt.title("Top Factors Correlated with Employee Attrition", fontsize=14, fontweight='bold')
plt.xlabel("Correlation with Attrition")
plt.ylabel("Feature")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualize specific important categorical factors
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
sns.boxplot(x='left', y='satisfaction_level', data=df, palette="pastel") # Changed Y to satisfaction_level
plt.title("Attrition vs Job Satisfaction")

plt.subplot(1,3,2)
sns.boxplot(x='left', y='average_montly_hours', data=df, palette="coolwarm") # Changed Y to average_montly_hours
plt.title("Attrition vs Monthly Hours")

plt.subplot(1,3,3)
sns.countplot(x='Work_accident', hue='left', data=df, palette="Set2") # Changed X to Work_accident
plt.title("Attrition by Work Accident Status")

plt.tight_layout()
plt.show()

# Step 7: (b) Projects vs Attrition
# The column representing number of projects is 'number_project'

if 'number_project' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='left', y='number_project', data=df, palette="Set1")
    plt.title("Attrition vs Number of Projects", fontsize=13, fontweight='bold')
    plt.xlabel("Attrition (0 = Stayed, 1 = Left)")
    plt.ylabel("Number of Projects")
    plt.grid(axis='y', alpha=0.6)
    plt.show()

    avg_projects = df.groupby('left')['number_project'].mean()
    print("=== (b) Average Number of Projects by Attrition Status ===")
    print(avg_projects)
else:
    print("⚠ Column 'number_project' not found in dataset. Please verify column name.")

# Step 8: Insights
print("\n=== Insights Summary ===")

# Identify top correlated factor
top_factor = top_factors.index[0]
print(f"💡 Strongest correlation with attrition: {top_factor}")

print("\nPossible Key Drivers:")
print("- Low job satisfaction and long overtime hours often lead to burnout and higher attrition.")
print("- Employees with lower monthly income or limited career growth tend to leave.")
print("- Those involved in too many or too few projects may feel disengaged or overworked.")

# Update the insights based on the actual column availability
if 'number_project' in df.columns:
    if avg_projects[1] > avg_projects[0]:
        print("\n📊 Employees with more projects are slightly more likely to leave.")
    else:
        print("\n📊 Employees with fewer projects are more likely to leave, possibly due to lack of engagement.")
else:
    print("\n📊 Cannot analyze project-related attrition without the 'number_project' column.")
```
